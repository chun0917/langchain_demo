```python
import whisper

model = whisper.load_model("base")
options = whisper.DecodingOptions(fp16 = False)
transcription = model.transcribe(video,language='zh')

text = transcription['text']
print(text)
```

```python
from datetime import timedelta
import os

def transcribe_audio(segments):
    VIDEO_FILENAME = "test2"

    for segment in segments['segments']:
        startTime = str(0)+str(timedelta(seconds=int(segment['start'])))+',000'
        endTime = str(0)+str(timedelta(seconds=int(segment['end'])))+',000'
        text = segment['text']
        segmentId = segment['id']+1
        segment = f"{segmentId}\n{startTime} --> {endTime}\n{text[1:] if text[0] == ' ' else text}\n\n"

        srtFilename = os.path.join(f"{VIDEO_FILENAME}.srt")
        with open(srtFilename, 'a', encoding='utf-8') as srtFile:
            srtFile.write(segment)

    return srtFilename


transcribe_audio(transcription)
```

```python
from langchain.document_loaders import SRTLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma

loader = SRTLoader("test.srt")
document = loader.load()

text_splitter = RecursiveCharacterTextSplitter(
  chunk_size=1000,
  chunk_overlap=20,
)

documents = text_splitter.split_documents(document)

for i in range(len(documents)):
    print(documents[i])

embeddings = OpenAIEmbeddings()
db = Chroma.from_documents(documents=documents,embedding=embeddings, persist_directory='db')

retriever = db.as_retriever()
```

```python
from langchain.chains import RetrievalQA
from langchain import OpenAI
import pyttsx3

qa = RetrievalQA.from_chain_type(llm=OpenAI(model='text-davinci-003'),
                       chain_type="stuff",
                       retriever=retriever,
                       return_source_documents=True)

query = "三大類模型是哪三大類"
result = qa({"query": query})
print(result["result"])
sound = pyttsx3.init()
sound.say(result["result"])
sound.runAndWait()
```
